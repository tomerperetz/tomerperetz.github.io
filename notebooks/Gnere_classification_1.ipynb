# Genre classification from musical and audio features

```python
%matplotlib inline
import sklearn
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import IPython.display as ipd
import scipy
from pathlib import Path
import seaborn as sns
```

```python
def plot_classes_pie(dataset, title=None):
    gdf = dataset.groupby('genre').count().reset_index()
    plt.pie(x=gdf['fname'], labels = gdf['genre'],autopct='%.0f%%')
    if title:
        plt.title(title)
    plt.show()
    print(f"n_samples: {dataset.shape[0]}")
```

```python
dataset = pd.read_csv("./audio/gtzan/manifest.csv")
plot_classes_pie(dataset, "Genres distribution")
```
